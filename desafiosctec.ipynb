```python
# Análise exploratória de dados do Titanic #

# Importar as bibliotecas#

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o Arquivo.csv com as dados da tabela #

titanic = sns.load_dataset('titanic')
titanic.to_csv('titanic.csv')

df=pd.read_csv('titanic.csv')
print(df.head())

# Manipulação dos dados para selecionar apenas variáveis de interesse, neste
# caso, Sexo e Sobreviventes do Naufrágio #

df_por_sexo = titanic.groupby('sex')['survived'].sum().reset_index()

print(df_por_sexo)


# Análise gráfica dos dados selecionados#

plt.figure(figsize=(8,6))

sns.barplot(data=df_por_sexo, x='sex', y='survived')

plt.show()

# Análise Simples dos Dados da Idade dos Passageiros

titanic = sns.load_dataset('titanic')
titanic.to_csv('titanic.csv')

plt.figure(figsize=(8,6))

sns.histplot(data=titanic, x='age')

plt.show()

##########################################
#Análise gráfica dos dados de voos ao longo dos meses e anos

# Heat Map dos Dados

voos= sns.load_dataset('flights')
voos = voos.pivot(index='month', columns='year', values='passengers')

plt.figure(figsize=(10,6))

sns.heatmap(voos, annot=True, fmt='.0f')

plt.show()


```
